```python
from ipyleaflet import Map, Circle
import requests
import json
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
```

```python
def search_by_adress(adress):
  url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={adress}&key={GOOGLE_API_KEY}"
  response = requests.get(url)
  data = json.loads(response.text)
  return data
```

```python
def make_circle(center, color, radius) -> Circle:
  circle = Circle()
  circle.location = center
  circle.radius = radius
  circle.color = color
  circle.fill_color = color
  return circle
```

```python
def get_point_google_response(response):
  return (response["geometry"]["location"]["lat"], response["geometry"]["location"]["lng"])
```

```python
def assign_color(score):
    color = [
        {
            "max": 2,
            "color": "yellow"
        },
        {
            "max": 4,
            "color": "orange"
        },
        {
            "max": 11,
            "color": "red"
        },
    ]
    for item in color:
        if score <= item["max"]:
            return item["color"]
    return "gray"
```

```python
df = pd.read_excel(r'GEOPORTAL PUNTOS CRITICOS AGOSTO2023 (1) (1).xlsx')
df
```

```python
df = df.fillna(0)
df
```

```python
df['SUMA_X'] = (df == 'X').sum(axis=1)
df = df[df['DISTRITO'] == 'COMAS']
df = df[["LUGAR", "SUMA_X", "TIPO DE DELITO", "MODALIDAD"]]
df
```

```python
puntos = []
for index, row in df.iterrows():
    data = search_by_adress(row["LUGAR"] + " COMAS LIMA")
    if len(data["results"]):
      puntos.append(get_point_google_response(data["results"][0]))
    else:
      puntos.append("a")
df["GEOLOCALIZACION"] = puntos
df = df[df['GEOLOCALIZACION'] != "a"]
df = df.reset_index()
df
```

```python
data = search_by_adress("distrito de comas")
comas = data["results"][0]
m = Map(center=(get_point_google_response(comas)), zoom=13)
comasArea = make_circle(get_point_google_response(comas), "gray", 5350)
m.add_layer(comasArea)
colors = []
for index, row in df.iterrows():
    score = row["SUMA_X"]
    assigned_color = assign_color(score)
    colors.append(assigned_color)
    m.add_layer(make_circle(row["GEOLOCALIZACION"], assigned_color, 500))
df["COLOR"] = colors
df
```
